```python
import pandas as pd
nba = pd.read_csv("nba.csv")
nba
```

```python
nba.dropna() ## drop all rows if any or all columns have NaN
```

```python
nba.dropna(how="any") ## same like default drop all rows if any or all columns have NaN
```

```python
nba.dropna(how="all") ## drop all rows if all its all columns have NaN
```

```python
nba.dropna(subset=['College']) ## drop all rows if all its column "College" have NaN
```

```python
nba.dropna(subset=['College', 'Salary']) ## drop all rows if all its column "College" or "Salary" have NaN [Its a Or relation not and]
```

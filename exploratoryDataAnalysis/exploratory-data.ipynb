## Conditional Selection
In performing exploratory data analysis, it is important to be able to select subsets of data to perform analysis or comparisons.

**Which player scored the most Points (PTS) Per Game? Here, we will return the entire row.**

```python
playerStats[playerStats["PTS"] == playerStats["PTS"].max()] ## There is only 1 row with maximum value of PTS
```

We will return specific column values.

##### **Further question: What team is the player from?**

```python
teamName = playerStats[playerStats["PTS"] == playerStats["PTS"].max()].Team
print(teamName)
```

##### **Which position is the player playing as?**

```python
position = playerStats[playerStats["PTS"] == playerStats["PTS"].max()].Pos
print(position)
```

##### **How many games did the player played in the season?**

```python
numberOfGames = playerStats[playerStats["PTS"] == playerStats["PTS"].max()].G
print(numberOfGames)
```

##### **Which player scored more than 20 Points (PTS) Per Game?**

```python
playersScoredMoreThan20 = playerStats[playerStats["PTS"] >= 20]
playersScoredMoreThan20
```

##### **Which player had the highest 3-Point Field Goals Per Game (3P) ?**

```python
highest3Point = playerStats[playerStats['3P'] == playerStats['3P'].max()]
highest3Point
```

##### **Which player had the highest Assists Per Game (AST) ?**

```python
highestAST = playerStats[playerStats['AST'] == playerStats['AST'].max()]
highestAST
```

### GroupBy() function
##### **Which player scored the highest (PTS) in the Los Angeles Lakers?**

```python
losAngelesLakers = playerStats.groupby('Team').get_group('LAL')
highestPts = losAngelesLakers[losAngelesLakers['PTS'] == losAngelesLakers['PTS'].max()]
highestPts
```

##### **Of the 5 positions, which position scores the most points?**

```python
ptsGroupByPos = playerStats.groupby('Pos')
ptsGroupByPos.PTS.describe()
```

```python
positions = ['C', 'PF', 'PG', 'SF', 'SG']
pos = playerStats[playerStats['Pos'].isin(positions)]
pos
```

```python
pos.groupby('Pos').PTS.describe()
```

```python
positions = pos['Pos'].unique()
maxPTS = float('-inf')
resultPos = None
for position in positions:
    currentPos = pos[pos['Pos'] == position]
    maxPTSByPos = currentPos[currentPos['PTS'] == currentPos['PTS'].max()].PTS.iloc[0]
    if maxPTSByPos > maxPTS:
        resultPos = position
        maxPTS = maxPTSByPos
print(f"Position {resultPos} have maximum PTS {maxPTS}")
```

```python
import pandas as pd

url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html = pd.read_html(url, header = 0)
playerStats = html[0]

## Data cleaning
playerStats = playerStats.drop(playerStats[playerStats['Age'] == 'Age'].index)
playerStats
```
